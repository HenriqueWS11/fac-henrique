### Energy Oscillations

#### Code to study the equations associated with the energy-dependent displacement $x_\epsilon$:

\begin{equation}
    x_\epsilon = \eta(s)\delta,
\end{equation}
where $\delta = \epsilon/E_0$, and $\eta$ is the single-valued function which satisfies:

\begin{equation}
    \eta'' = -K_x(s)\eta + G(s),
\end{equation}
and which is $L$-periodic in $s$.

##### Field free section

$$G = 0;\,\, K_x = 0$$

\begin{equation}
    \implies \eta' = cte
\end{equation}

##### Pure quadrupoles

$$G(s) = 0;$$ $$K_x(s) = \pm|K_x|$$

Focusing quadrupole ($K_x > 0$):

\begin{equation}
    \eta = a\cdot cos(\sqrt{K_x}s + \vartheta)
\end{equation}

Defocusing quadrupole ($K_x < 0$):

\begin{equation}
    \eta = a\cdot cosh(\sqrt{-K_x}s + \vartheta)
\end{equation}

#### Pure dipole (flat bending magnet)

$K_1 = 0 \implies K_x = G^2$

\begin{equation}
    \implies \eta'' = -G^2\eta + G
\end{equation}

Which leads to solutions of the type:

\begin{equation}
    \eta = Asin(Gs) + Bcos(Gs) + 1/G
\end{equation}

This corresponds to a sinusoidal oscillation around $1/G$ - the radius of curvature $\varrho_s$ of the design orbit - with a "restoring force" proportional to $G²$.

Since $\eta$ is $L$-periodic, we have the further constraints:

$$\eta(L) = \eta(0);$$  $$\eta'(L) = \eta'(0),$$

whose further development leads to:

$$Asin(GL) + Bcos(GL) = B \implies A sin(GL) = B (1 - cos(GL))$$
$$AGcos(GL) - BGsin(GL) = AG \implies B sin(GL) = A (cos(GL) - 1)$$

These two conditions together can impose restrictions on the possible values of G, which will be related with A, B and L.

Plotting possible solutions with $1/G = 82.44$ m:

```python
import numpy as np
from matplotlib import pyplot as plt   

A = [0, 2, 3, 3]
B = [3, 2, 1, 0]

G = 1/20 # 1/m
s = np.linspace(0, 500, 1000) # [m]

def f(s, A, B, G):
    return A * np.sin(G * s) + B * np.cos(G * s) + 1/G

plt.figure(figsize=(10, 6))
for i in range(len(A)):
    plt.plot(s, f(s, A[i], B[i], G), label=f'A = {A[i]}, B = {B[i]}')
plt.hlines(0, s[0], s[-1], color='darkblue', linestyle='--', label='Ideal orbit')
plt.hlines(1/G, s[0], s[-1], color='black', linestyle='--', label='1/G') # Add horizontal line for 1/G
plt.xlabel('s [m]')
plt.ylabel(r'$\eta$(s) [m]')
plt.ylim(-5, 30)
plt.xlim(0, s[-1])
plt.title('A.sin(G.s) + B.cos(G.s) + 1/G for different A and B')
plt.grid()
plt.legend(loc='lower right')
plt.show()

delta = 0.001
plt.figure(figsize=(10, 6))
for i in range(len(A)):
    plt.plot(s, delta*f(s, A[i], B[i], G), label=f'A = {A[i]}, B = {B[i]}')
plt.hlines(0, s[0], s[-1], color='darkblue', linestyle='--', label='Ideal orbit')
plt.hlines(delta*1/G, s[0], s[-1], color='black', linestyle='--', label=r'$\delta/G$ = $\delta\varrho_s$') # Add horizontal line for 1/G
plt.xlabel('s [m]')
plt.ylabel(r'$x_{\epsilon}$(s) [m]')
plt.ylim(-5*delta, 30*delta)
plt.xlim(0, s[-1])
plt.title(r'Typical $x_{\epsilon}$ = $\eta(s)\delta$ at a Dipole (with $\neq$ init. cond.) for $\delta=0.001$')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
